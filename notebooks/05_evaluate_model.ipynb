```python
# ============================================
# 05_evaluate_model.ipynb
# Evaluate model predictions and produce
# executive-facing player value table.
# ============================================

import pandas as pd
import numpy as np
import os
os.chdir(r"C:\AI Projects\NFL Player Rating")
```

```python
# -----------------------------
# Load model output + features
# -----------------------------
engineered = pd.read_csv("data/engineered_dataset.csv")
model_output = pd.read_csv("data/model_output.csv")

print("Engineered:", engineered.shape)
print("Model output:", model_output.shape)
```

```python
# -----------------------------
# Merge engineered + predictions
# -----------------------------
df = pd.merge(
    engineered,
    model_output[["player_name", "season", "predicted_cap_percent"]],
    on=["player_name", "season"],
    how="inner"
)

print("Merged:", df.shape)
df.head()
```

```python
# -----------------------------
# Compute difference
# -----------------------------
df["diff"] = df["cap_percentage"] - df["predicted_cap_percent"]
```

```python
# -----------------------------
# Verdict classification
# -----------------------------
def classify(diff):
    if diff > 0.02:
        return "Overpaid"
    elif diff < -0.02:
        return "Underpaid"
    else:
        return "Fair Value"

df["verdict"] = df["diff"].apply(classify)
```

```python
# -----------------------------
# Executive-facing table
# -----------------------------
final = df[[
    "player_name", "position", "team", "season",
    "custom_rating",
    "cap_percentage",
    "predicted_cap_percent",
    "diff",
    "verdict"
]].sort_values("diff")

final.head(20)
```

```python
# -----------------------------
# Save final output
# -----------------------------
final.to_csv("data/final_player_values.csv", index=False)
print("Saved final_player_values.csv")
```
