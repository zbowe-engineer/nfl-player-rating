```python
# ============================================
# 01_explore_nfl_data.ipynb
# Load and explore raw NFL data using nfl_data_py
# ============================================

import pandas as pd
import nfl_data_py as nfl
import os
os.chdir(r"C:\AI Projects\NFL Player Rating")
```

```python
# -----------------------------
# Choose seasons to download
# -----------------------------
years = list(range(2019, 2025))
years
```

```python
# -----------------------------
# Load season-level data
# -----------------------------
season_df = nfl.import_seasonal_data(years)
print("Season data shape:", season_df.shape)

season_df.head()
```

```python
# -----------------------------
# Load weekly-level data
# -----------------------------
weekly_df = nfl.import_weekly_data(years)
print("Weekly data shape:", weekly_df.shape)

weekly_df.head()
```

```python
# -----------------------------
# Basic cleaning
# -----------------------------
season_df = season_df.drop_duplicates()
weekly_df = weekly_df.drop_duplicates()

print("Season after cleaning:", season_df.shape)
print("Weekly after cleaning:", weekly_df.shape)
```

```python
# -----------------------------
# Save cleaned raw data
# -----------------------------
season_df.to_csv("data/season_raw.csv", index=False)
weekly_df.to_csv("data/weekly_raw.csv", index=False)

print("Saved season_raw.csv and weekly_raw.csv")
```
