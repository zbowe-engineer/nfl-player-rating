```python
# -----------------------------
# Load merged dataset
# -----------------------------
df = pd.read_csv("data/merged_dataset.csv")
print("Loaded merged dataset:", df.shape)

df.head()
```

```python
# -----------------------------
# Per-game stats
# -----------------------------
df["yards_per_game"] = (
    df["passing_yards"] +
    df["rushing_yards"] +
    df["receiving_yards"]
) / df["games_played"]

df["td_rate"] = (
    df["passing_tds"] +
    df["rushing_tds"] +
    df["receiving_tds"]
) / df["games_played"]

df["touches_pg"] = (
    df["attempts"] +
    df["carries"] +
    df["targets"]
) / df["games_played"]

df["snaps_pg"] = df["snaps"] / df["games_played"]
```

```python
# -----------------------------
# Efficiency metrics
# -----------------------------
df["yards_per_attempt"] = df["passing_yards"] / df["attempts"].replace(0, np.nan)
df["yards_per_carry"] = df["rushing_yards"] / df["carries"].replace(0, np.nan)
df["yards_per_reception"] = df["receiving_yards"] / df["targets"].replace(0, np.nan)

df["efficiency"] = (
    df["yards_per_attempt"].fillna(0) * 0.4 +
    df["yards_per_carry"].fillna(0) * 0.3 +
    df["yards_per_reception"].fillna(0) * 0.3
)
```
